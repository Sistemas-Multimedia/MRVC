# Computation of the energy of the wavelet coefficients



```python
import pywt # pip install pywavelets
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np

WAVELET = 'bior3.5' #'db5'
K=2 #level

frameb = np.zeros((64,64))
coeffs = np.array(frameb)
coeffs[32,32] = 255
```

### Calculating Energy HH

```python
LL=frameb
LH=frameb
HL=frameb
HH=coeffs

plt.figure()

plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.imshow(LL)

plt.subplot(2, 2, 2)
plt.imshow(LH)

plt.subplot(2,2,3)
plt.imshow(HL)

plt.subplot(2,2,4)
plt.imshow(HH)


imageHH = pywt.idwt2((LL,(LH,HL,HH)), WAVELET, mode='per')
plt.figure()
plt.imshow(imageHH)

energyHH = np.sum(imageHH*imageHH)
```

### Calculating Energy HL

```python
LL=frameb
LH=frameb
HL=coeffs
HH=frameb

plt.figure()

plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.imshow(LL)

plt.subplot(2, 2, 2)
plt.imshow(LH)

plt.subplot(2,2,3)
plt.imshow(HL)

plt.subplot(2,2,4)
plt.imshow(HH)


imageHL = pywt.idwt2((LL,(LH,HL,HH)), WAVELET, mode='per')
plt.figure()
plt.imshow(imageHL)

energyHL = np.sum(imageHL*imageHL)
```

### Calculating Energy LH

```python
LL=frameb
LH=coeffs
HL=frameb
HH=frameb

plt.figure()

plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.imshow(LL)

plt.subplot(2, 2, 2)
plt.imshow(LH)

plt.subplot(2,2,3)
plt.imshow(HL)

plt.subplot(2,2,4)
plt.imshow(HH)


imageLH = pywt.idwt2((LL,(LH,HL,HH)), WAVELET, mode='per')
plt.figure()
plt.imshow(imageLH)

energyLH = np.sum(imageLH*imageLH)
```

### Calculating Energy LL

```python
if K == 1: 
    LL=coeffs
    LH=frameb
    HL=frameb
    HH=frameb

    plt.figure()

    plt.subplot(2, 2, 1) # (rows, columns, panel number)
    plt.imshow(LL)

    plt.subplot(2, 2, 2)
    plt.imshow(LH)

    plt.subplot(2,2,3)
    plt.imshow(HL)

    plt.subplot(2,2,4)
    plt.imshow(HH)


    imageLL = pywt.idwt2((LL,(LH,HL,HH)), WAVELET, mode='per')
    plt.figure()
    plt.imshow(imageLL)

    energyLL = np.sum(imageLL*imageLL)
elif K==2:
    LH=frameb
    HL=frameb
    HH=frameb
    
    frameb2 = np.zeros((32,32))
    coeffs2 = np.array(frameb2)
    coeffs2[16,16] = 255
    
    LL2=frameb2
    LH2=frameb2
    HL2=frameb2
    HH2=coeffs2
    
    imageHH2 = pywt.idwt2((LL2,(LH2,HL2,HH2)), WAVELET, mode='per')
    imageLL = pywt.idwt2((imageHH2,(LH,HL,HH)), WAVELET, mode='per')
    energyHH2 = np.sum(imageLL*imageLL)
    
    plt.figure()
    plt.imshow(imageLL)
    
    LL2=frameb2
    LH2=frameb2
    HL2=coeffs2
    HH2=frameb2
    
    imageHL2 = pywt.idwt2((LL2,(LH2,HL2,HH2)), WAVELET, mode='per')
    imageLL = pywt.idwt2((imageHL2,(LH,HL,HH)), WAVELET, mode='per')
    energyHL2 = np.sum(imageLL*imageLL)
    
    plt.figure()
    plt.imshow(imageLL)
    
    LL2=frameb2
    LH2=coeffs2
    HL2=frameb2
    HH2=frameb2
    
    imageLH2 = pywt.idwt2((LL2,(LH2,HL2,HH2)), WAVELET, mode='per')
    imageLL = pywt.idwt2((imageLH2,(LH,HL,HH)), WAVELET, mode='per')
    energyLH2 = np.sum(imageLL*imageLL)
    
    plt.figure()
    plt.imshow(imageLL)
    
    LL2=coeffs2
    LH2=frameb2
    HL2=frameb2
    HH2=frameb2
    
    imageLL2 = pywt.idwt2((LL2,(LH2,HL2,HH2)), WAVELET, mode='per')
    imageLL = pywt.idwt2((imageLL2,(LH,HL,HH)), WAVELET, mode='per')
    energyLL2 = np.sum(imageLL*imageLL)
    
    plt.figure()
    plt.imshow(imageLL)
    
    
```

### Calculating Subband gains

```python
gainHH = energyHH/energyHH
gainHL = energyHL/energyHH
gainLH = energyLH/energyHH
if K==1:
    gainLL = energyLL/energyHH
    print("Energy HH, HL, LH, LL")
    print(energyHH, energyHL, energyLH, energyLL)
    print("gainHH=",gainHH)
    print("gainHL=",gainHL)
    print("gainLH=",gainLH)
    print("gainLL=", gainLL)
elif K==2:
    gainHH2 = energyHH2/energyHH2
    gainHL2 = energyHL2/energyHH2
    gainLH2 = energyLH2/energyHH2
    gainLL2 = energyLL2/energyHH2
    print("Energy HH, HL, LH, LL")
    print(energyHH, energyHL, energyLH, energyLL)
    print("Energy HH2, HL2, LH2, LL2")
    print(energyHH2, energyHL2, energyLH2, energyLL2)
    print("gainHL=",gainHL)
    print("gainLH=",gainLH)
    print("gainHL2=", gainHL2)
    print("gainLH2=", gainLH2)
    print("gainLL2=", gainLL2)
    
```
