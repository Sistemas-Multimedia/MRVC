# Computation of the energy of the wavelet coefficients



```python
import pywt # pip install pywavelets
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np


frameb = np.zeros((64,64))
coeffs = np.array(frameb)
coeffs[32,32] = 255
```

## Calculating Energy HH

```python
LL=frameb
LH=frameb
HL=frameb
HH=coeffs

plt.figure()

plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.imshow(LL)

plt.subplot(2, 2, 2)
plt.imshow(LH)

plt.subplot(2,2,3)
plt.imshow(HL)

plt.subplot(2,2,4)
plt.imshow(HH)


imageHH = pywt.idwt2((LL,(LH,HL,HH)), 'db5', mode='per')
plt.figure()
plt.imshow(imageHH)

energyHH = np.sum(imageHH*imageHH)
```

**Calculating Energy HL

```python
LL=frameb
LH=frameb
HL=coeffs
HH=frameb

plt.figure()

plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.imshow(LL)

plt.subplot(2, 2, 2)
plt.imshow(LH)

plt.subplot(2,2,3)
plt.imshow(HL)

plt.subplot(2,2,4)
plt.imshow(HH)


imageHL = pywt.idwt2((LL,(LH,HL,HH)), 'db5', mode='per')
plt.figure()
plt.imshow(imageHL)

energyHL = np.sum(imageHL*imageHL)
```

**Calculating Energy HL

```python
LL=frameb
LH=coeffs
HL=frameb
HH=frameb

plt.figure()

plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.imshow(LL)

plt.subplot(2, 2, 2)
plt.imshow(LH)

plt.subplot(2,2,3)
plt.imshow(HL)

plt.subplot(2,2,4)
plt.imshow(HH)


imageLH = pywt.idwt2((LL,(LH,HL,HH)), 'db5', mode='per')
plt.figure()
plt.imshow(imageLH)

energyLH = np.sum(imageLH*imageLH)
```

**Calculating Energy HL

```python
LL=coeffs
LH=frameb
HL=frameb
HH=frameb

plt.figure()

plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.imshow(LL)

plt.subplot(2, 2, 2)
plt.imshow(LH)

plt.subplot(2,2,3)
plt.imshow(HL)

plt.subplot(2,2,4)
plt.imshow(HH)


imageLL = pywt.idwt2((LL,(LH,HL,HH)), 'db5', mode='per')
plt.figure()
plt.imshow(imageLL)

energyLL = np.sum(imageLL*imageLL)
```

**Calculating Subband gains

```python
gainHH = energyHH/energyHH
gainHL = energyHL/energyHH
gainLH = energyLH/energyHH
gainLL = energyLL/energyHH

print("Energy HH, HL, LH, LL")
print(energyHH, energyHL, energyLH, energyLL)
print("gainHH=",gainHH)
print("gainHL=",gainHL)
print("gainLH=",gainLH)
print("gainLL=", gainLL)
```
